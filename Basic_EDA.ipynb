## Classification Model

```python
df = pd.read_csv('PH_posts_data.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['createdAt']).dt.date
```

Generally anything above 200 upvotes on Product Hunt has a chance of becoming product of the day and 200-300 upvotes is considered a good lauch.

* Hence 200-500 votes for a launch is great.
* Anything above 500 can be considered an outlier 
* If a launch gets between 75 and 200 votes, its a good launch
* If it gets anywhere between 25-75, its average
* Less than 25 -> low

```python
def categorize_votes(vote):
    if vote > 500:
        return 'outlier'
    elif vote > 200:
        return 'great'
    elif vote > 75:
        return 'good'
    elif vote > 25:
        return 'average'
    else:
        return 'low'
```

```python
df['vote_category'] = df['votesCount'].apply(categorize_votes)
```

```python
df['description']
```

```python
from langchain_community.embeddings import HuggingFaceInferenceAPIEmbeddings
```

```python
HF_token = os.getenv('HUGGINGFACEHUB_API_TOKEN')
```

```python
embeddings = HuggingFaceInferenceAPIEmbeddings(
    api_key=HF_token, model_name="BAAI/bge-small-en-v1.5"
)

query_result = embeddings.embed_query(df['description'][0])
query_result[:3]
```

```python
emb = embeddings.embed_documents(df['description'].to_list())
```
